**Affichage des titres de tous les events sur la page principale**


```python
def scrap_titles(browser):
   div = browser.find_element_by_class_name('feed-events-bucket__content')
   links = div.find_elements_by_tag_name('img')  
   titres = []
   for  l in links:
        titres.append(l.get_attribute('alt'))
   return titres
scrap_titles(browser)
```

**Affichage des titres**

```python
def scrap_titles(links):
    titres = []
    for  I in links:
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        titre = soup.find('h1', 'listing-hero-title')
        titres.append(titre.text)
    return titres
titres=scrap_titles(links)
titres
```

```python
len(titres)
```

**Affichage des organismes qui ont crée les évenements**

```python
import re
def scrap_organises(links):
    organises = []
    for  I in links:
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        organise=soup.find('a', 'js-d-scroll-to listing-organizer-name text-default')
        try:
            organises.append(re.sub(r'\s', '', organise.text))
        except:
            organises.append('None')
        pass
    return organises
organises=scrap_organises(links)
organises
```

```python
len(organises)
```

**Affichage de la date des evenets**

```python
def scrap_dates(links):
    dates = []
    for  I in links:
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        date=soup.find('p', 'js-date-time-second-line')
        try:
            dates.append(date.text)
        except:
            organises.append('dates multiples')
        pass
    return dates
dates=scrap_dates(links)
dates
```

```python
len(dates)
```

**Affichage d'un seule nombre d'abbones le problèmes si au niveau dela charge de la page pour récupérer le nombre d'abonnées**

```python
driver=webdriver.Chrome(executable_path='C:\chromedriver.exe', chrome_options=chrome_options)
driver.get('https://www.eventbrite.fr/e/billets-week-end-de-transformation-entrepreneur-55544991565?aff=ebdssbdestsearch')
driver.maximize_window() 
driver.implicitly_wait(20)
abbones=[]
abbone=driver.find_element_by_xpath("//div[@class='follow-organizer__follow-count']")
#nettyage de donneé
a=abbone.text.split(' ')
abonne=a[0]
abonne
abbones.append(abonne)
abbones
```

```python
def scrap_followers(links):
    followers = []
    for  I in links:
        driver.maximize_window() 
        driver.implicitly_wait(20)
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        follower=soup.find('div', 'follow-organizer__follow-count')
        try:
            followers.append(follower)
        except:
            organises.append('0')
        pass
        
    return followers
followers=scrap_followers(links)
followers
```

```python
len(followers)
```

```python
import re
def scrap_location(links):
    locations = []
    for  I in links:
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        location=soup.find({'p'},{'class':'event-details__data'},'p')
        location=re.sub(r'\s', '', location.text)
        try:
            locations.append(location)
        except:
            locations.append('null')
        pass
    return locations
locations=scrap_location(links)
locations
```

```python
len(locations)
```

```python
import re
def scrap_prix(links):
    prix = []
    for  I in links:
        response = requests.get(I)
        soup = BeautifulSoup(response.content, 'lxml')
        montant=soup.find('div', 'js-panel-display-price')
        montant=re.sub(r'\s', '', montant.text)
        prix.append(montant)
    return prix
prix=scrap_prix(links)
prix
```

```python
len(prix)
```
